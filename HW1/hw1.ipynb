Problem 1: Test the MATLAB image functions to read, display, and write images. Use
buckeyes_gray.bmp and buckeyes_rgb.bmp from the class materials webpage. [2 pts]

```python
import skimage
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
grayIm = io.imread("buckeyes_gray.bmp")
plt.imshow(grayIm, cmap='gray')
plt.axis("image")
io.imsave("buckeyes_gray.jpg", grayIm)
rgbIm = io.imread("buckeyes_rgb.bmp")
plt.imshow(rgbIm)
plt.axis("image")
io.imsave("buckeyes_rgb.jpg", rgbIm)
```

Problem 2: Read and convert the rgb image to grayscale using the NTSC conversion formula via the MATLAB function rgb2gray. Display your image to verify the result. [1 pt]

```python
grayIm = skimage.color.rgb2gray(rgbIm)
plt.imshow(grayIm, cmap='gray')
plt.axis("image")
```

Problem 3: Test more fully by creating, writing, and reading a checker-board image. [2 pts]

```python
zBlock = np.zeros((10, 10))
oBlock = np.ones((10, 10)) * 255
temp1 = np.hstack((zBlock, oBlock))
temp2 = np.hstack((oBlock, zBlock))
pattern = np.vstack((temp1, temp2))
checkerIm = np.tile(pattern, (5, 5))
io.imsave('checkerIm.bmp', np.uint8(checkerIm))
Im = io.imread('checkerIm.bmp')
io.imshow(Im)
```
